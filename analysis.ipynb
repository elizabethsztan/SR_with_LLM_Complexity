# Comparison of outputted complexities of Qwen2.5-Instruct 0.5B vs 7B model

Experiment to see the outputted complexities of two LLMs and two datasets.

**Simple** dataset: $$y = x_1^2 + 2*x_2 + 4.5$$\
**Complex** dataset: $$y = 2.7*\exp(x_1 - 0.2*x_2)*x_2^{0.04} + (1 - x_1)$$


Only ran for 10 iterations, population size 20, number of populations 1.

Example complexities given to the LLM: "x1 + x2 + C has complexity 3, C * sin(x1) has complexity 4, sin(exp(sin(x1))) has complexity 10"

```python
import json
import matplotlib.pyplot as plt
```

```python
qwen05b_simple_no = 1764848118
qwen05b_complex_no = 1764850842
```

```python
qwen05b_data = {}

with open(f"experimental_results/qwen05b/simple_expression/llm_complexity_log_{qwen05b_simple_no}.json", "r") as f:
    qwen05b_data["llm_simple"] = json.load(f)

with open(f"experimental_results/qwen05b/simple_expression/standard_complexity_log_{qwen05b_simple_no}.json", "r") as f:
    qwen05b_data["standard_simple"] = json.load(f)

with open(f"experimental_results/qwen05b/complex_expression/llm_complexity_log_{qwen05b_complex_no}.json", "r") as f:
    qwen05b_data["llm_complex"] = json.load(f)

with open(f"experimental_results/qwen05b/complex_expression/standard_complexity_log_{qwen05b_complex_no}.json", "r") as f:
    qwen05b_data["standard_complex"] = json.load(f)
```

```python
qwen7b_simple_no = 1764952747
qwen7b_complex_no = 1764955961
```

```python
qwen7b_data = {}

with open(f"experimental_results/qwen7b/simple_expression/llm_complexity_log_{qwen7b_simple_no}.json", "r") as f:
    qwen7b_data["llm_simple"] = json.load(f)

with open(f"experimental_results/qwen7b/simple_expression/standard_complexity_log_{qwen7b_simple_no}.json", "r") as f:
    qwen7b_data["standard_simple"] = json.load(f)

with open(f"experimental_results/qwen7b/complex_expression/llm_complexity_log_{qwen7b_complex_no}.json", "r") as f:
    qwen7b_data["llm_complex"] = json.load(f)

with open(f"experimental_results/qwen7b/complex_expression/standard_complexity_log_{qwen7b_complex_no}.json", "r") as f:
    qwen7b_data["standard_complex"] = json.load(f)
```

```python
fig, axes = plt.subplots(4, 2, figsize=(10, 12))
num_bins = 30

#simple
axes[0, 0].hist(qwen05b_data["standard_simple"], bins = num_bins)
axes[0, 0].set_title(f"Qwen2.5-0.5B-Instruct\n \n Standard simple: {len(qwen05b_data["standard_simple"])}")

axes[0, 1].hist(qwen7b_data["standard_simple"], bins = num_bins)
axes[0, 1].set_title(f"Qwen2.5-7B-Instruct\n \n Standard simple: {len(qwen7b_data["standard_simple"])}")

axes[1, 0].hist(qwen05b_data["llm_simple"], bins = num_bins)
axes[1, 0].set_title(f"LLM simple: {len(qwen05b_data["llm_simple"])}")

axes[1, 1].hist(qwen7b_data["llm_simple"], bins = num_bins)
axes[1, 1].set_title(f"LLM simple: {len(qwen7b_data["llm_simple"])}")

#complex

axes[2, 0].hist(qwen05b_data["standard_complex"], bins = num_bins)
axes[2, 0].set_title(f"Standard complex: {len(qwen05b_data["standard_complex"])}")

axes[2, 1].hist(qwen7b_data["standard_complex"], bins = num_bins)
axes[2, 1].set_title(f"Standard complex: {len(qwen7b_data["standard_complex"])}")

axes[3, 0].hist(qwen05b_data["llm_complex"], bins = num_bins)
axes[3, 0].set_title(f"LLM complex: {len(qwen05b_data["llm_complex"])}")

axes[3, 1].hist(qwen7b_data["llm_complex"], bins = num_bins)
axes[3, 1].set_title(f"LLM complex: {len(qwen7b_data["llm_complex"])}")


plt.tight_layout()
plt.show()
```
